```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
t = np.arange(0, 2, 0.01)
seno = np.sin(2*np.pi*t)
coseno = np.cos(2*np.pi*t)
```

## lineplot

```python
sns.lineplot(t, seno)
sns.lineplot(t, coseno)
```

```python
df = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
df
```

```python
df["target"] = load_iris().target
df
```

## scatterplot

```python
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)",
                hue="target")
```

```python
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)",
                hue="target", style="target")
```

## histplot

```python
sns.histplot(data=df, x="sepal length (cm)")
```

```python
sns.histplot(data=df, y="sepal length (cm)")
```

```python
sns.histplot(data=df, x="sepal length (cm)", bins=30)
```

```python
sns.histplot(data=df, x="sepal length (cm)", kde=True)
```

## barplot

```python
tips = sns.load_dataset("tips")
tips
```

```python
ax = sns.barplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

## boxplot

```python
ax = sns.boxplot(y=tips["total_bill"])
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

## violinplot

```python
ax = sns.violinplot(y=tips["total_bill"])
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

## heatmap

```python
matriz_confusion = np.array([[22,6,2],
                             [6,20,4],
                             [5,1,24]])
```

```python
ax = sns.heatmap(matriz_confusion)
```

```python
ax = sns.heatmap(matriz_confusion, annot=True)
```
