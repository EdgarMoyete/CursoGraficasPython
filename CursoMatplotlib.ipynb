```python
import numpy as np
import matplotlib.pyplot as plt
```

## plot

```python
plt.plot(([2,4,6,8,10]))
plt.grid()
plt.show()
```

```python
x = np.arange(10)
x
```

```python
y1 = x * x
y1
```

```python
plt.plot(x, y1)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["f(x)"])
plt.grid()
plt.show()
```

```python
y2 = x * x * x
y2
```

```python
plt.plot(x, y1) # x^2
plt.plot(x, y2) # x^3
plt.title("f(x) = x^2, g(x)=x^3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["f(x)", "g(x)"])
plt.grid()
plt.show()
```

```python
t = np.arange(0, 2, 0.01)
seno = np.sin(2 * np.pi * t)
coseno = np.cos(2 * np.pi * t)
```

```python
plt.plot(t, seno)
plt.plot(t, coseno)
plt.title("f(x) = sin(x), g(x)=cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["f(x)", "g(x)"])
plt.grid()
# plt.axis([0, 1.5, 0, 0.75])
plt.show()
```

## subplot

```python
plt.subplot(2, 1, 1)
plt.plot(t, seno)
plt.title("f(x) = sen(x), g(x)=cos(x)")
plt.ylabel("y")
plt.legend(["f(x)"])
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, coseno)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["g(x)"])
plt.grid()
```

## pcolormesh, imshow, matshow

```python
matriz_confusion = np.array([[22, 6, 2],
                             [6, 20, 4],
                             [5, 1, 24]])
```

```python
# seal 8 segundos con frecuencia de muestreo de 128Hz
# suma de 3 seales seno

fs = 128
time = np.arange(0, 8, 1/fs)
x1 = np.sin(2*np.pi*2*time) # seno 2hz
x2 = np.sin(2*np.pi*5*time) # seno 5hz
x3 = np.sin(2*np.pi*8*time) # seno 8hz
sig = x1 + x2 + x3
```

```python
plt.plot(time, sig)
plt.xlabel("tiempo [s]")
plt.ylabel("amplitud")
plt.grid()
plt.show()
```

```python
from scipy.signal import spectrogram
```

```python
f, t, a = spectrogram(sig, fs=fs, window="hann", nperseg=fs, noverlap=116)
```

### pcolormesh

```python
plt.pcolormesh(t, f, a)
plt.xlabel("tiempo [s]")
plt.ylabel("frecuencia [hz]")
plt.show()
```

```python
plt.pcolormesh(matriz_confusion)
plt.show()
```

## imshow

```python
plt.imshow(a)
plt.show()
```

```python
plt.imshow(matriz_confusion)
plt.show()
```

## matshow

```python
plt.matshow(a)
plt.show()
```

```python
plt.matshow(matriz_confusion)
plt.show()
```
